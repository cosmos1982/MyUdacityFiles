**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
## 拆分导演, 原来数据中单部电影多名导演是用’|’隔开的
tmp = movie_data['director'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('director') 
movie_data_split = movie_data[['original_title', 'revenue']].join(tmp)


## 用index来记录导演的排序
top10_director = pd.Series(movie_data_split['director'].value_counts()[:10].index, name='director').reset_index() 

## 用inner的merge来代替筛选, 并进行二重排序
sort_data = movie_data.merge(top10_director, on='director')\
                      .sort_values(['index','revenue_adj'],ascending=[True,False])

## 用groupby得到目标数据
target_data = sort_data.groupby('director').head(3) 

## 作图, 将导演设为颜色维度hue, 达到分别作图的效果
fig = plt.figure(figsize=(15, 6)) 
ax = sb.barplot(data=target_data, x='original_title', y='revenue_adj', hue='director', dodge=False, palette="Set2")
plt.xticks(rotation = 90)
plt.ylabel('Revenue')
plt.xlabel('Original Title')
```

```python
#此处排行前3没有说明是什么排行前3，因此取popularity的前三。
#modify: 将director中被|分隔的导演名称进行split
tmp = movie_data['director'].str.split('|', expand=True).stack().reset_index(level=1, drop=True).rename('director') 
movie_data_split = movie_data[['original_title', 'revenue', 'popularity']].join(tmp)

movie_counts = movie_data_split.groupby('director').size()
top_directors = movie_counts.sort_values(ascending=False)[0:11];

#经过观察top_directors中存在unknown，unknown的情况下分析意义不大，因此对数据进行手工调整一下
top_directors = top_directors[[0] + list(range(2,11))];
print(top_directors)

top1_data = []
top2_data = []
top3_data = []

for director in top_directors.index:
    movies_of_director = movie_data_split[movie_data_split['director'] == director].sort_values('popularity', ascending=False)[0:3]
    top1_data.append(movies_of_director['revenue'].values[0])
    top2_data.append(movies_of_director['revenue'].values[1])
    top3_data.append(movies_of_director['revenue'].values[2])

revenue_data = pd.DataFrame(data = {'top1': top1_data, 'top2': top2_data, 'top3': top3_data}, index=top_directors.index.values)
#print(revenue_data)

fig = plt.figure(figsize = [15, 5])
ax = fig.add_subplot(1, 1, 1)
revenue_data.plot.bar(ax = ax)
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_rotation(30)

#modify，补充坐标和标题
plt.title("Revenue of most popular movies top3 of most productive directors top10")
plt.xlabel("Director name (with decending order of movie couns)")
plt.ylabel("Top3 movie revenue")
```

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
movie_data['release_year_month'] = movie_data['release_date'].map(lambda x:100*x.year + x.month)
selected_movie = movie_data[(movie_data['release_year_month'] >= 196801) & (movie_data['release_year_month'] <= 201506)]
selected_movie.head()
```

```python
# 绘制散点图
selected_movies_counts = selected_movie.groupby('release_year_month').size().reset_index(name='count')

selected_movies_counts.head()
fig = plt.figure(figsize = [15, 5]);
ax = fig.add_subplot(1, 1, 1);
sb.regplot('release_year_month', 'count', data = selected_movies_counts);
ax.set_xlim([196801, 201507]);
ax.set_xticks(list(range(196801, 201507, 400)));
xtick_label = ["{}-{:0>2d}".format(x//100, x%100) for x in ax.get_xticks()];
ax.set_xticklabels(xtick_label);
plt.xlabel('')
```

```python
#任意绘制年度数据，观察一年内每个月份的数据变化
def show_figure(year, ax):
    sb.regplot('release_year_month', 'count', data = selected_movies_counts)
    ax.set_xlim([year*100+1, year*100+12])
    ax.set_xticks(list(range(year*100+1, year*100+13, 1)))
    xtick_label = ["{}-{:0>2d}".format(x//100, x%100) for x in ax.get_xticks()]
    ax.set_xticklabels(xtick_label)
    for tick in ax.xaxis.get_major_ticks():
        #tick.label.set_fontsize(10)
        tick.label.set_rotation(30)
    ax.set_title('Monthly movie count statics in ' + str(year))
    plt.xlabel('')
    plt.ylabel('Movie count')

fig = plt.figure(figsize = [16, 10]);
ax = fig.add_subplot(2, 2, 1);
show_figure(2009, ax)

ax = fig.add_subplot(2, 2, 2);
show_figure(2011, ax)

ax = fig.add_subplot(2, 2, 3);
show_figure(2012, ax)

ax = fig.add_subplot(2, 2, 4);
show_figure(2014, ax)

plt.subplots_adjust(hspace=0.3, wspace=0.2)
```

<h3>任务3.4结论</h3><br>
1.整体上来看，从1968年到2015年，电影的数量是逐渐上升的；<br>
2.单独分析2009, 2011, 2012, 2014年的数据,可以看出电影数量在暑期档最多。

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
selected_comedy_movies = movie_data[(movie_data['release_year_month'] >= 196801) & (movie_data['release_year_month'] <= 201506) & (movie_data['genres']=='Comedy')]
selected_comedy_movies_count = selected_comedy_movies.groupby('release_year_month').size().reset_index(name = 'comedy_count')
selected_comedy_movies_count1 = selected_comedy_movies_count.set_index('release_year_month')

selected_drama_movies = movie_data[(movie_data['release_year_month'] >= 196801) & (movie_data['release_year_month'] <= 201506) & (movie_data['genres']=='Drama')]
selected_drama_movies_count = selected_drama_movies.groupby('release_year_month').size().reset_index(name = 'drama_count')
selected_drama_movies_count1 = selected_drama_movies_count.set_index('release_year_month')

#print(selected_drama_movies_count.head())

#绘制comedy数量图
fig = plt.figure(figsize = [16, 15]);
ax = fig.add_subplot(3, 1, 1);
selected_comedy_movies_count1.plot(ax = ax, linestyle='dashed', marker="o")
ax.set_title('Movies count of Comedy During 1968 - 2015-06')
ax.set_xlim([196801, 201507]);
ax.set_xticks(list(range(196801, 201507, 400)));
xtick_label = ["{}-{:0>2d}".format(x//100, x%100) for x in ax.get_xticks()];
ax.set_xticklabels(xtick_label);
plt.xlabel('')

#绘制drama数量图
ax = fig.add_subplot(3, 1, 2);
selected_drama_movies_count1.plot(ax = ax, linestyle='dashed', marker="o")
ax.set_title('Movies count of Drama During 1968 - 2015-06')
ax.set_xlim([196801, 201507]);
ax.set_xticks(list(range(196801, 201507, 400)));
xtick_label = ["{}-{:0>2d}".format(x//100, x%100) for x in ax.get_xticks()];
ax.set_xticklabels(xtick_label);
plt.xlabel('')

#绘制comedy和drama一起的条形图
result = pd.merge(selected_comedy_movies_count,selected_drama_movies_count, on='release_year_month', how = 'outer')
result = result.fillna(0).set_index('release_year_month')
result.sort_index(inplace = True)

ax = fig.add_subplot(3, 1, 3);
result.plot.bar(ax=ax);
ax.set_xticks(ax.get_xticks()[::6]);
ax.set_xticklabels(["{}-{:0>2d}".format(x//100, x%100) for x in result.index[::6]]);

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_rotation(30)
ax.set_title('Movies count of Comedy and Drama During 1968 - 2015-06')
plt.xlabel('')
```
